### Fetching Listing URLs:en.

```python
from bs4 import BeautifulSoup
import requests
import time
from requests import get

headers = {'User-Agent': 'Mozilla/5.0'}

url = 'https://sfbay.craigslist.org/search/zip#search=1~gallery~0~0'
page = requests.get(url, headers)
soup = BeautifulSoup(page.content, 'html.parser')

```

#### Extracting the first 250 unique listing URLs and saving them to a list. 
I will use soup.find_all() function to find all the elements with tag 'a' and then get the href link from those tags.

#### Printing the list to screen.

```python
#Selecting the element which contains the individsual item link
listings = soup.select('li.cl-static-search-result')
#creating an empty list to store the links
listing = []

#creating a for loop which would iterate 250 times to fetch the individsual links
for i in range(0,254):
    ind_listing = listings[i]
    listing_link = ind_listing.select_one('a').get('href')
    #Appending the links in a list
    listing.append(listing_link)
    
#Printing out the list of first 250 final links
print(listing)

```

### Explanation:
Selecting the element which contains the individsual item link. Then, creating an empty list to store the links. Next, creating a for loop which would iterate 250 times to fetch the individsual links
The index to get the desired output starts at 4 which fetches the first individsual item on the website and runs for next 250 times. Within the for loop, append the links in a list. Printing out the list of first 250 final links

```python
#Creating a function to store the html file
def saveString(html, filename="test.html"):
	try:
		file = open(filename, "w", encoding='utf-8')
		file.write(str(html))
		file.close()
	except Exception as ex:
		print('Error: ' + str(ex))
        
time.sleep(10)

#calling the function to store the files
for j, k in zip(listing, filename):
    try :
        url2 = j
        page2 = requests.get(url2)
        soup2 = BeautifulSoup(page2.content, 'html.parser')
        saveString(soup2, k)
        print('File saved sucessfully')
    except Exception as ex:
        print('Error: ' + str(ex))
   
   
```

### Extracted Information:

For each HTML file, use `BeautifulSoup` to parse the file content.

Extract and print the following details:

Title: The title of the listing.

URL of first image (if an image exists):  The URL of the displayed image.  It can be found in the `src` attribute of `<img>`

Description: The full description text of the listing.

Post ID: Usually found at the bottom of the page or within the page's HTML structure.

Posted Date: The date when the listing was originally posted.

Last Updated Date: The date when the listing was last updated.

```python
# Loop through each file in the directory
time.sleep(5)

for filename in os.listdir(directory):
    # Check if the file ends with .html
    if filename.endswith(".html"):
        # Construct the full file path
        filepath = os.path.join(directory, filename)
        # Read file to string
        with open(filepath, 'r', encoding='utf-8') as file:
            html = file.read()
            soup3 = BeautifulSoup(html, "html.parser")
            title = soup3.select('title')
            print("Title :",title)
            image = [img['src'] for img in soup3.select('img[src]')]
            print("Image:", image)
            posting_body = soup3.select_one('section#postingbody ')
            desc = posting_body.get_text(strip = True)
            print('Description:', desc)
            post_id = soup3.select_one('div.postinginfos > p.postinginfo')
            print('Post_ID :', post_id.text)
            post_date = soup3.select('p.postinginfo.reveal')
            print('Posting date:', post_date[0].text)
            print('Update date:', post_date[2].text)
           
           

         
            
    
            
```

### Automating the Login Process

Using Python and appropriate libraries like requests, simulate the login process

```python
time.sleep(10)
headers = {'User-agent': 'Mozilla/5.0'}

url4 = 'https://theoldreader.com/users/sign_in'
page4 = requests.get(url4, headers=headers)
soup4 = BeautifulSoup(page4.content, 'html.parser')

print(soup4.prettify())
```

```python
input = soup4.select_one('#new_user input[name=utf8]')
utf8 = input.get('value')
print(utf8)

input2 = soup4.select_one('#new_user input[name = authenticity_token]')
authenticity_token = input2.get('value')
print(authenticity_token)

input3 = soup4.select_one('#new_user input[name = commit]')
commit = input3.get('value')
print(commit)
```

Create a session object to maintain your login state across multiple requests.

```python
# Always pause between two requests.
time.sleep(5) # 5s

# An open session carries the cookies and allows you to make post requests
session = requests.session()

```

Prepared a payload with my login credentials and other necessary form data identified from the login page and the network analysis.                    Sent a POST request to the login form’s action URL to log in, using the session object.

```python
res = session.post('https://theoldreader.com/users/sign_in', 
                        data = {'utf8' : utf8,
                                'authenticity_token' : authenticity_token,
                                'user[login]' : 'priyal98.sf@gmail.com', 
                                'user[password]' : 'Priyalvijay',
                                 'commit' : commit},
                        # headers = dict(referer = 'https://www.neopets.com/login/'),
                        timeout = 20)
```
